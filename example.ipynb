# Computing deformations that generate conjugate points along Kolmogorov flows on the torus

A Kolmogorov flow is a geodesic in the space $\mathrm{Diff}_{\mu}(\mathbb T^2)$ of volume-preserving diffeomorphisms of the torus, whose velocity field $X = \mathrm{sgrad}\, \psi$ is generated by a stream function of the form $\psi(x,y) = \cos(mx)\cos(ny)$ for integers $m$ and $n$. The goal is to find vector fields along these geodesics that make the Misiolek index negative. When the Misiolek index can be made negative, the index form is also negative for some sufficiently large time and there is a conjugate point along the geodesic for the considered values of $m$ and $n$.

The vector fields that we look for are generated by a spatial deformation $f(x,y)$ in the following way: $Y = \mathrm{sgrad}\, g$, where $g(t,x,y) = \sin(t/T)f(x,y)$, and we search for minimizers $f$ of unit $p$-homogenous Sobolev norm. The Misiolek index can be represented by a quadratic operator acting on the spatial deformation $f$ via the $\dot H^p$ inner product. The minimal value of the operator on deformations of unit $\dot H^p$-norm is given by the minimal eigenvalue of its matrix representation, and the associated eigenvector gives the minimizer $f$. When the eigenvalue is negative, then there is a conjugate point.

The matrix representation is computed with respect to the coordinate system defined by the Fourier coefficients of $f(x,y) = \sum_{j}\sum_k a_{jk} \cos(jx+ky)$, where the sum is for $j = 0$ and $k = 0,...,N$, and $j=1,...,N$ and $k= -N,..., N$.

We minimize the Misiolek index for different values of $m$ and $n$ and different orders $p$ of the homogeneous Sobolev inner product. The Misiolek index operator is compact for $p\geq 2$ and bounded for $p\geq 3$. For $p\geq 2$, we see a characteristic ”V-shape” appearing, i.e. the deformation that minimizes the Misiolek criterion is a a perturbation of $\cos(x)$.

```python
p_list = [0, 1, 2, 3]
```

```python
m, n = 2, 1

deformation_list = []
misiolek_index_list = []
for p in p_list:
    deformation, misiolek_index = minimize_misiolek_index(m, n, N, p)
    deformation_list.append(deformation)
    misiolek_index_list.append(misiolek_index)
deformation_list = np.stack(deformation_list)

plot_deformation(deformation_list, misiolek_index_list)
```

```python
m, n = 3, 2

deformation_list = []
misiolek_index_list = []
for p in p_list:
    deformation, misiolek_index = minimize_misiolek_index(m, n, N, p)
    deformation_list.append(deformation)
    misiolek_index_list.append(misiolek_index)
deformation_list = np.stack(deformation_list)

plot_deformation(deformation_list, misiolek_index_list)
```

In the diagonal case $m = n$, due to the symmetry in the variables $x$ and $y$, another minimizer is given by a perturbation of $\cos(y)$, the symmetric perturbation obtained by exchanging $x$ and $y$. In the $(2,2)$ case, we can see both minimizers appearing separately as well as mixed, as in the case $p = 1$.

```python
m, n = 2, 2

deformation_list = []
misiolek_index_list = []
for p in p_list:
    deformation, misiolek_index = minimize_misiolek_index(m, n, N, p)
    deformation_list.append(deformation)
    misiolek_index_list.append(misiolek_index)
deformation_list = np.stack(deformation_list)

plot_deformation(deformation_list, misiolek_index_list)
```

To always obtain a perturbation of $\cos(x)$, we constrain the solution to have Fourier coefficient $a_{01} = 0$.

```python
m, n = 2, 2

deformation_list = []
misiolek_index_list = []
for p in p_list:
    deformation, misiolek_index = minimize_misiolek_index(m, n, N, p, constrained=True)
    deformation_list.append(deformation)
    misiolek_index_list.append(misiolek_index)
deformation_list = np.stack(deformation_list)

plot_deformation(deformation_list, misiolek_index_list)
```
